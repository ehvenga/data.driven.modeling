```python
import numpy as np
```

```python
data = np.loadtxt('hw1Q1.txt')
data
```

```python
l1_norms = np.sum(np.abs(data), axis=0)
l2_norms = np.sqrt(np.sum(np.square(data), axis=0))
linf_norms = np.max(np.abs(data), axis=0)
```

```python
largest_l1 = np.argmax(l1_norms)
largest_l2 = np.argmax(l2_norms)
largest_linf = np.argmax(linf_norms)
```

```python
print(f"Largest L1 norm is in column {largest_l1 + 1} with a value of {l1_norms[largest_l1]}")
print(f"Largest L2 norm is in column {largest_l2 + 1} with a value of {l2_norms[largest_l2]}")
print(f"Largest L∞ norm is in column {largest_linf + 1} with a value of {linf_norms[largest_linf]}")
```
